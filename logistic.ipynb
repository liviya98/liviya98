```python
import numpy as np
import matplotlib.pyplot as plt
# Generate A Simple dataset
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
Y = (4 + 3 * X + np.random.randn(100 , 1) > 0).astype(int)
#Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
#   Intitalize weights
theta = np.random.randn(2, 1)
#add bias term to X
X_b = np.c_[np.ones((100 ,1)) ,X]
#Learning rate
learning_rate = 0.01
#Number of iterations
n_iteration = 1000
#Gradient Descent Without explicit epochs
for _ in range(n_iteration):
  #Calculate predicted probabilities
  Y_proba = sigmoid(X_b.dot(theta))
  #Update Weights using the gradient
  gradients =  X_b.T.dot(Y_proba - Y)
  theta = theta - learning_rate * gradients
# Plot the dataset and desicion boundary
plt.scatter(X,Y)
plt.xlabel("x")
plt.ylabel("y")
# Plot the desicion boundary
x_decision_boundary = np.linspace(np.min(X), np.max(X) , 100)
X_decision_boundary= np.c_[np.ones ((100 ,1)), x_decision_boundary]
Y_decision_boundary = sigmoid(X_decision_boundary.dot(theta))
plt.plot(x_decision_boundary, Y_decision_boundary, "r-" , label="Decision Boundary")

plt.legend()
plt.show()
```
